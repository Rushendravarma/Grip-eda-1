# FACET HISTORAMS

```python
g_h = sns.FacetGrid(rishi,row="Region",col="Segment",palette="Set1")
g_h = g_h.map(plt.hist,"Profit")
```

```python
g= sns.FacetGrid(rishi,row="Region",col="Segment",hue="Ship Mode",palette="Set1")
kws = dict(s=40,linewidth=0.6,edgecolor="black")
g= g.map(plt.scatter,"Sales","Profit",**kws)
```

```python
g= sns.FacetGrid(rishi,row="Region",col="Segment")
kws = dict(s=40,linewidth=0.6,edgecolor="blue")
g= g.map(plt.scatter,"Sales","Profit",**kws)
g.set(xlim=(15,70),ylim=(0,22))
```

```python
g= sns.FacetGrid(rishi,row="Region",col="Segment",hue="Ship Mode")
kws = dict(s=40,linewidth=0.6,edgecolor="blue")
g= g.map(plt.scatter,"Sales","Profit",**kws)
g.set(xlim=(15,70),ylim=(0,22))
```

```python
g= sns.FacetGrid(rishi,row="Region",col="Segment",hue="Ship Mode",palette="Set1")
kws = dict(s=40,linewidth=0.6,edgecolor="blue")
g= g.map(plt.scatter,"Sales","Profit",**kws)
g.set(xlim=(0,80),ylim=(0,30))
for ax in g.axes.flat:
    ax.plot((0,80),(0,30),c="orange",ls="--")
g.add_legend()
```
